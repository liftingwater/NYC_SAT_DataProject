```python
import pandas as pd
import numpy as np
```

```python
#Begin reading/ examining data

data_files = ["ap_2010.csv", "class_size.csv", "demographics.csv", "graduation.csv", "hs_directory.csv", "math_test_results.csv", "SAT_results.csv"]

data = {}
for f in data_files: 
    d = pd.read_csv("SAT_Data/" + f)
    data[f.replace(".csv", "")] = d
```

```python
def examine_data(i): 
    if i == 0: 
        #view column headers
        for k, v in data.items(): 
            print("\n"+k+"\n")
            print(v.columns.values)
    elif i == 1: 
        #View dataframe heads (5)
        for k, v in data.items(): 
            print("\n"+str(k)+"\n")
            print(v.head(5))
    else: 
        pass
```

```python
examine_data(1)
```

```python
#Making DBN columns for dfs that dont have one
data["class_size"]["DBN"] = data["class_size"].apply(lambda x: "{0:02d}{1}".format(x["CSD"], x["SCHOOL CODE"]), axis=1)

data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```
